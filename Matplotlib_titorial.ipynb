## Science des données
</div>

Colab vous permet de tirer pleinement parti des bibliothèques populaires Python pour analyser et visualiser des données. La cellule de code ci-dessous utilise <strong>numpy</strong> pour générer des données aléatoires et <strong>matplotlib</strong> pour les visualiser. Pour modifier le code, cliquez simplement sur la cellule.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
x = [1,2,3,4,5,6,7,8]
y = [4,1,3,6,1,3,5,2]
size = [100,500,100,500,100,500,100,500]

plt.scatter(x,y,s=size)
#  "Les attribus du graphique"
plt.title('Nuage de points avec Matplotlib')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('ScatterPlot_02.png')
plt.show()
```

**l'interpretation:** on remarque un mouvement aleatoire, les variables ne sont pas correles entre eux

la p_value est un seuil pour prendre la decisiondans un test statistique


```
"Création de variables"
x = [1,2,3,4,5,6,7,8]
y = [4,1,3,6,1,3,5,2]
y2 = [1, 4, 9, 16, 25,36,49,64]
"Création du graphique de point"
plt.title('Avec des couleurs cest mieux!!!')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y, c = 'brown',marker = 'D',label="courbe brown")
plt.scatter(x, y2, c='purple',marker = '*', label="courbe purple")
plt.annotate('point atypique', xy = (1, 4), xytext = (2, 22),arrowprops = {'facecolor': 'pink',
'shrink': 0.1})
plt.legend()
plt.grid()
plt.show()
```

les variables x et y ont un comportement aleatoire, aloers qu'il y'a un comportement quadratique entre x et Y2        
   pour bien rediges un commentaire jeter un coup d'oeil sur les rapports de l'HCP, l'fme, l'BMCE

```
"Création du graphique"

plt.title('la courbe sous Matplotlib')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y, 'r*-',c = 'red',label="courbe rouge")
plt.plot(x, y2, 'gd-',c = 'green',label="courbe verte")
plt.legend()
plt.show()
```

le quartile est la valeur de la variable, sous laquelle on a 25% de la population                         
pour voir la repartition de la population et la point abberants(atypique) et les supprimer ou les corriger

quand les donnees ne sont pas coherant, on cherche les points aberrants, la boite a mostache donne une idee sur la qualite de la database

```
# Simulation des données

np.random.seed(19680801)
x1 = np.random.rand(50) * 100
x2 = np.ones(25) * 50
x_high = np.random.rand(10) * 100 + 100
x_low = np.random.rand(10) * -100
data = np.concatenate((x1, x2, x_high, x_low))


# Boxplot avec Matplotlib

plt.title('BOXPLOT avec Matplotlib')
green_diamond = dict(markerfacecolor='orange', marker='*')
plt.boxplot(data, flierprops= green_diamond, vert=False)
plt.show()
```

on utulise un dictionnaire pour definir la coleur et le style par defaut pour personnaliser le rapport et creer un beau rapport okkk

```
# Structure des données

labels = 'groupe 1', 'groupe 2', 'groupe 3', 'groupe 4'
sizes = [215, 130, 245, 210]
# Mise en forme du graphique

colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = ( 0, 0, 0.1, 0) # juste le premier groupe

# graphique
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)
plt.show()
```

**explode : ** pour marquer un groupe et le faire sortir par rapport au autre groupe                              
** '%1.1f%': ** le nombre de nb apres la virgule                                            
NB : le circle est urulise pour visualiser 5 groupe au max au dela de 5on utilise un histogramme
